### Pandas to Numpy

```python
dfSPY=pd.read_csv("data/SPY.csv")

print(dfSPY.head())
```

To port from PD (which is essentially a wrapper for NP) to NP:

```python
nd1=dfSPY.values
print(nd1[0:5])
```

Slicing

```python
print(nd1[0:3][0:3])
```

```python
a=np.random.normal(0,50,size=(5,20))
plt.plot(a)
plt.show()
```

```python
np.random.seed(123)
a=np.random.randint(-100,100,size=(5,20))
plt.plot(a)
plt.show()
```
