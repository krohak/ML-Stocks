```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def symbol_to_path(symbol, base_dir="data"):
    """Return CSV file path given ticker symbol."""
    return os.path.join(base_dir, "{}.csv".format(str(symbol)))


def get_data(symbols, dates):
    """Read stock data (adjusted close) for given symbols from CSV files."""
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols:  # add SPY for reference, if absent
        symbols.insert(0, 'SPY')

    for symbol in symbols:
        df_temp = pd.read_csv(symbol_to_path(symbol), index_col='Date',
                parse_dates=True, usecols=['Date', 'Adj Close'], na_values=['nan'])
        df_temp.rename(columns={'Adj Close': symbol}, inplace=True)
        df = df.join(df_temp)
        if symbol == 'SPY':  # drop dates SPY did not trade
            df = df.dropna(subset=["SPY"])

    return df
```

## Rolling Mean

```python
dates = pd.date_range('2010-01-01', '2010-12-31')
df = get_data(symbols=[],dates=dates)
```

```python
df.head()
```

```python
ax = df['SPY'].plot(title="SPY Rolling Mean",label="SPY")
```

```python
rm_SPY = df['SPY'].rolling(window=20).mean()
```

```python
rm_SPY.plot(label="Rolling Mean",ax=ax)
```

```python
ax.set_xlabel('Time')
ax.set_ylabel('Price')
ax.legend(loc='upper left')
plt.show()
```

## Bollinger Bands

```python
rstd_SPY = df['SPY'].rolling(window=20).std()
```

```python
upper_band = rm_SPY + 2*rstd_SPY
lower_band = rm_SPY - 2*rstd_SPY
```

```python
upper_band.plot(label="Upper Band",ax=ax)
lower_band.plot(label="Lower Band",ax=ax)
```

```python
ax.set_xlabel('Time')
ax.set_ylabel('Price')
ax.legend(loc='upper left')
plt.show()
```

## Daily Returns

How much did the price go up or down on a particular day.

We use df.values to convert to np because if we try to subtract dataframes, pd will try to match indices

```python
df_int = df.copy()
df_int[1:] = (df[1:] - df[:-1].values) / df[:-1].values
df_int.iloc[0] = 0
df_int.head()
```

```python
df_int.plot()
plt.show()
```

Alternatively, we use `df.shift(n)` to use the dataframe itself to compute the difference:

```python
df_int = (df / df.shift(1)) - 1
df_int.iloc[0] = 0
df_int.head()
```

```python
df_int.plot()
plt.show()
```

## Cumulative Returns

Percentage change starting from the first date till t

```python
df_int = (df / df.iloc[0]) - 1
df_int.head()
```

```python
df_int.plot()
plt.show()
```
